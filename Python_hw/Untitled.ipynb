```python
class MyList(list):
    def __iter__ (self):
        return my_iter(self)


class my_iter:
    def __init__(self,lst):
        self.lst = lst
        self.i = 0
    def __next__(self):
        if self.i < len(self.lst):
            if self.lst[self.i] % 3 == 0:
                res = True
            else:
                res = False
            self.i += 1
            return(res)
        else:
            raise StopIteration
```

```python
lst = [1,2,3,4,5,6]
for x in MyList(lst):
    print(x)
```
